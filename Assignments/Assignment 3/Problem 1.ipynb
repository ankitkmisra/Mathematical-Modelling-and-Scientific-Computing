# Problem 1

## Problem Statement

Solve the following differential equation using odeint on python and plot the graph of y vs x:

dy/dx = cos(x) + x^2

Take y(0) = 0 and x = linspace(0,10,1000).

## Solution:

```python
from scipy.integrate import odeint
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
def model(y, x):
    dydx = np.cos(x) + x*x
    return dydx

y0 = 0

x = np.linspace(0, 10, 1000)

y = odeint(model, y0, x)
```

```python
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
