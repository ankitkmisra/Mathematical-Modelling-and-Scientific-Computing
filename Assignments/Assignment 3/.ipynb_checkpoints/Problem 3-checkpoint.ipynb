# Problem 3

## Problem Statement

Model a simple vertical spring-mass system and plot its position and velocity as a function of
time.

The time should span from 0 to 10 with 10000 sample points in between.

Take k = 2.5, m = 1.5, g = 9.8 (all in SI units) and the initial position and velocity to both be 0.

## Solution:

```python
from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt
```

```python
def model(y, t, k, m, g):
    x, v = y
    dxdt = v
    dvdt = m*g - k*x
    return ([dxdt, dvdt])
```

```python
x0 = 0
v0 = 0

t = np.linspace(0, 10, 10000)

y = odeint(model, [x0, v0], t, args=(2.5, 1.5, 9.8))
```

```python
plt.figure()
plt.plot(t, y[:, 0])
plt.xlabel('Time')
plt.ylabel('Displacement')
plt.show()
```

```python
plt.figure()
plt.plot(t, y[:, 1])
plt.xlabel('Time')
plt.ylabel('Velocity')
plt.show()
```
