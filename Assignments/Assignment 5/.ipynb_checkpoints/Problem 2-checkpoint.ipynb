# Problem 2 - Damping in RLC Circuits

## Problem Statement:

You are given a simple series RLC circuit, with no power supply. The specifications are :

R = 2Ω, L = 1H, C = xF

1. You have to find the value of the unknown capacitance x, such that you achieve a critically damped circuit (Note that once you have the x value for the critically damped case, just changing it would take you to either the overdamped or the underdamped case, depending on the direction you make the change in).

2. For the initial conditions q0 = 50.0C, i0 = 0.0A, and the time t array taken as np.arange(0, 20, 0.05) produce the following 6 plots:
    1. The first two plots should correspond to the ‘q vs t’ and ‘i vs t’ plots respectively, for the overdamped case.
    2. The next two plots should correspond to the ‘q vs t’ and ‘i vs t’ plots respectively, for the critically damped case.
    3. The last two plots should correspond to the ‘q vs t’ and ‘i vs t’ plots respectively, for the underdamped case.

Note: For each of the 6 plots, mention which damping case it represents, along with the x value used, in the title.

## Solution:

### Part 1:

We have, for the circuit, Li'' + Ri' + i/C = 0, and thus, for critical damping, R^2 = 4L/C. This gives us x = 1.

Thus we have:
1. For overdamping, C > 1F.
2. For critical damping, C = 1F.
3. For underdamping, C < 1F.

### Part 2(A):

```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.integrate import odeint
%matplotlib inline
```

```python
def model(state, t, R, L, C):
    q, i = state
    dqdt = i
    didt = -(i*R + q/C)/L
    return [dqdt, didt]

R = 2.0
L = 1.0
C = 2.0
q0 = 50.0
i0 = 0.0

t = np.arange(0.0, 20.0, 0.05)

y = odeint(model, [q0, i0], t, args=(R, L, C))

plt.figure(figsize=[10, 6])
plt.plot(t, y[:, 0])
plt.title('Overdamping (x = 2.0)')
plt.xlabel('Time, t')
plt.ylabel('Charge, q')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=[10, 6])
plt.plot(t, y[:, 1])
plt.title('Overdamping (x = 2.0)')
plt.xlabel('Time, t')
plt.ylabel('Current, i')
plt.grid(True)
plt.show()
```

### Part 2(B):

```python
C = 1.0

y = odeint(model, [q0, i0], t, args=(R, L, C))

plt.figure(figsize=[10, 6])
plt.plot(t, y[:, 0])
plt.title('Critical Damping (x = 1.0)')
plt.xlabel('Time, t')
plt.ylabel('Charge, q')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=[10, 6])
plt.plot(t, y[:, 1])
plt.title('Critical Damping (x = 1.0)')
plt.xlabel('Time, t')
plt.ylabel('Current, i')
plt.grid(True)
plt.show()
```

### Part 2(C):

```python
C = 0.2

y = odeint(model, [q0, i0], t, args=(R, L, C))

plt.figure(figsize=[10, 6])
plt.plot(t, y[:, 0])
plt.title('Underdamping (x = 0.2)')
plt.xlabel('Time, t')
plt.ylabel('Charge, q')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=[10, 6])
plt.plot(t, y[:, 1])
plt.title('Underdamping (x = 0.2)')
plt.xlabel('Time, t')
plt.ylabel('Current, i')
plt.grid(True)
plt.show()
```
