# Problem 1 (Dirichlet Boundary Conditions)

## Problem Statement

For a rectangular plate of length l = 2 metres (horizontal) and width w = 1 metre (vertical), simulate the steady state temperature contours for the following boundary conditions:

1. Temperature at the top edge is (100 + 50sin(2 * pi * x/l)) degree Celsius (constant in time)
2. At all the other edges, the temperature is maintained at 100 degree Celsius

## Solution:

```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import clear_output

%matplotlib inline
```

```python
maxIter = 1000

dX = 0.1
dY = 0.1
lenX = 2 + dX
lenY = 1 + dY

Tbottom = 100
Tleft = 100
Tright = 100

Tguess = 0
```

```python
colorInterpolation = 100
colorMap = plt.cm.jet

X, Y = np.meshgrid(np.arange(0, lenX, dX), np.arange(0, lenY, dY))
```

```python
T = np.empty((int(lenY/dY), int(lenX/dX)))
T.fill(Tguess)

for i in range(0, int(lenX/dX)):
    T[(int(lenY/dY) - 1):, i] = 100 + 50*np.sin(2*np.pi*i*dX/lenX)
T[:1, :] = Tbottom
T[:, int(lenX/dX - 1):] = Tright
T[:, :1] = Tleft
```

```python
tmp = 2/(dX**2) + 2/(dY**2)

print("Iterating!")

for iteration in range(maxIter):
    for j in range(1, int(lenX/dX) - 1):
        for i in range(1, int(lenY/dY) - 1):
            T[i,j] = ((T[i+1,j] + T[i-1,j])/(dY**2) + (T[i,j+1] + T[i,j-1])/(dX**2))/tmp

print("Iteration finished.")
```

```python
plt.figure(figsize=[14, 7])
plt.style.use('dark_background')
plt.suptitle("Temperature Contours", fontsize=20)
plt.contourf(X, Y, T, colorInterpolation, cmap=colorMap)
plt.xlabel("X nodes", fontsize=15)
plt.ylabel("Y nodes", fontsize=15)
plt.colorbar()
```
