# Problem 3 (Langton's Ant)

## Problem Statement

An ant starts out on a random grid containing dark/black (0) and light/white (1) cells, and then follows the following set of rules:

1. If the ant is on a black square, it turns right and moves forward one unit.
2. If the ant is on a white square, it turns left and moves forward one unit.
3. When the ant leaves a square, it inverts the color.

Make a 7*7 grid where you simulate the motion of Langton’s ant.

The output should be an animated matrix, similar to the example code for Conway’s Game of Life. You do not need to “show an ant”. The evolution of the grid is sufficient. You will need to store the ant's position in a way that you can use it again.

Hint 1: Use np.random.randint(2, size=(n,n))) to create the random grid of numbers where n is the dimensionality of the grid.

Hint 2: You only need to change the animate function in the Game of Life code.

Hint 3: To get a random initial point for the ant use: indices =  np.random.randint(0, high=max_index, size=2)

Use boundary conditions such that the board wraps around itself (top and bottom row are joined and leftmost and rightmost columns are joined). The ant is essentially living on a toroid!

## Solution:

```python
%matplotlib notebook
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.animation import FuncAnimation
```

The following function `animation_frame` defines the changes to be made to the grid at each step, using the rules stated above.

```python
def animation_frame(i):
    global n, board, ant_x, ant_y, ant_dir
    
    if board[ant_y, ant_x] == 0:
        ant_dir = (ant_dir+1)%4
        board[ant_y, ant_x] = 1
    else:
        ant_dir = (ant_dir-1)%4
        board[ant_y, ant_x] = 0
    
    if ant_dir == 0:
        ant_x = (ant_x+1)%n
    elif ant_dir == 1:
        ant_y = (ant_y+1)%n
    elif ant_dir == 2:
        ant_x = (ant_x-1)%n
    else:
        ant_y = (ant_y-1)%n
        
    im.set_array(board)
    return im
```

The following code runs the animation for Langton's ant.

```python
fig, ax = plt.subplots()

n = 7
board = np.random.randint(2, size=(n,n))
ant_x = np.random.randint(n)
ant_y = np.random.randint(n)
ant_dir = np.random.randint(4)

im = plt.imshow(board, animated=True, cmap='gray')

animation = FuncAnimation(fig, animation_frame, frames=500, interval=500, blit=True)
plt.show()
```
